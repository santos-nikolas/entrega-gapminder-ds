## Passo 1 - importando bibliotecas e o dataset

```python
import matplotlib.pyplot as plt
import pandas as pd

# Carregando o CSV
df = pd.read_csv("ddf-datapoints-literacy-rate-adult-by-geo-gender-time.csv")
df
```

## Passo 2 - filtrando dados apenas para o Brasil, China e índia e plotando gráfico
### Informações do gráfico:
<ul>
    <li><b>1º dimensão:</b> x - Ano</li>
    <li><b>2º dimensão:</b> y - taxa de alfabetização de adultos</li>
    <li><b>3º dimensão:</b> tamanho da bolinha - país (em ordem crescente de tamanho: Brasil, China, Índia)</li>
    <li><b>4º dimensão:</b> cor - gênero (roxo feminino e amarelo masculino)</li>
</ul>

```python
# Filtra o DataFrame para incluir apenas Brasil, China, Índia e EUA
df = df[df['geo'].isin(['bra', 'chn', 'ind'])]

# Redefine o índice da DataFrame
df = df.reset_index(drop=True)

x = df['time']
y = df['literacy_rate_adult']

# Define o tamanho da bolinha para cada país
size_map = {'bra': 500, 'chn': 1000, 'ind': 1500}
size = df['geo'].map(size_map)
color = df['gender']  # Cor baseada no gênero

# Cria o scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, s=size, c=color, alpha=0.5)

# Configura o gráfico
plt.xlabel('Ano')
plt.ylabel('Taxa de Alfabetização de Adultos')
plt.title('Taxa de Alfabetização de Adultos por Região e Gênero ao Longo do Tempo')

# Adiciona uma barra de cores para o gênero
plt.colorbar(label='Gênero (0: Feminino, 1: Masculino)')

# Adiciona rótulos para cada bolinha
for i, txt in enumerate(df['geo']):
    plt.text(x[i], y[i], txt, ha='center', va='center', fontsize=8)

# Exibe o gráfico
plt.show()
```
